# Vietnamese MCQA — Run on Kaggle / FPT
Cell duy nhat: clone repo, cai dat, chay inference, download ket qua.


```python
import os, sys, subprocess

# === CONFIG ===
os.environ['MCQA_BACKEND'] = 'vllm'       # vllm | hf | unslo
os.environ['MCQA_MODEL_ID'] = 'Qwen/Qwen3.5-4B'  # Qwen3.5-4B | Qwen3.5-9B | google/gemma-4-E4B-it
os.environ['MCQA_BATCH_SIZE'] = '8'

DATA_DIR = '.'        # thu muc chua file .json test
OUTPUT_DIR = './output'
REPO = 'vsh-nlp-agent'

# === SETUP ===
if not os.path.isdir(REPO):
    subprocess.run(['git', 'clone', 'https://github.com/HiImSunny/vsh-nlp-agent'], check=True)
os.chdir(REPO)

# Second: pip deps (transformers moi cho Qwen3.5, vllm)
subprocess.run(['pip', 'install', '-q',
    'vllm>=0.8.0',
    'git+https://github.com/huggingface/transformers.git',
    'accelerate', 'sentencepiece', 'tqdm'
], check=True)

# === INFERENCE ===
os.makedirs(OUTPUT_DIR, exist_ok=True)
subprocess.run([
    sys.executable, 'src/run.py',
    '--data-dir', DATA_DIR,
    '--output-dir', os.path.abspath(OUTPUT_DIR),
], check=True)

# === DOWNLOAD ===
from IPython.display import FileLink
display(FileLink(f'{OUTPUT_DIR}/pred.csv'))
```
